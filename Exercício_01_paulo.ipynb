##1.3 Carregamento do DataFrame na API RestCountries

```
def reqCountriesAll():
  res = requests.get('https://restcountries.eu/rest/v2/all')
  if res.status_code == 200:
    # return res.json()
    _restcountries = res.json()
    pd_restcountries = pd.DataFrame.from_dict(_restcountries)
    return pd_restcountries
  else:
    print(f'Erro {res.status_code} : "{res.reason}"')

pd_restcountries =  reqCountriesAll()
pd_restcountries.head() 
```

```
list_restcountries = list(pd_restcountries.columns)
ar_restcountries = []
for x in list_restcountries:
  ar_restcountries.append(x)
print(ar_restcountries)
print(len(ar_restcountries))
```
