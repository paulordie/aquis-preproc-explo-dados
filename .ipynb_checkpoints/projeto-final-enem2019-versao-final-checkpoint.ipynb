# Pré-processamento dos Dados

```python
def basic_clean_df(data_frame, columns_to_remove_list, organize_by_column):
    
    #UPPERCASE + Remove spaces to change for '_' from columns names
    data_frame.columns = data_frame.columns.str.upper().str.replace(' ', '_')
    
    #Remove line brokes
    data_frame.columns = data_frame.columns.str.replace('\n', '')
    
    #Remover Colunas que foram passadas por parâmetro
    for i in columns_to_remove_list:
        if i in data_frame.columns:
            data_frame = data_frame.drop(columns=i)
    
    #Varrer linhas e colocar em Uppercase nos dados das colunas
    try:
        columns = data_frame.columns
        index = 0
        for i in columns:
            if i in data_frame:
                if (data_frame.dtypes[i]==object):
                    data_frame[i] = data_frame[i].str.upper()
                    string_ = data_frame[i]
                    if string_[0] == ' ':
                        aux = string[1:]
            index += 1
    except:
        print('error')
        
    
    
    #Ordenar Dataframe pela coluna passada como parâmetro 
    data_frame = data_frame.sort_values(organize_by_column)
    
    return data_frame
```

### **Como está o desempenho dos alunos brasileiros, segundo o ENEM 2019? Média das notas do brasil e média por região?**

```python
def media(row):
    return (row['NU_NOTA_CN'] + row['NU_NOTA_CH'] + row['NU_NOTA_LC'] + row['NU_NOTA_MT'])/4

def regiao(row):
    if row['CO_MUNICIPIO_ESC'][0] == '1':
        return 'NORTE'
    elif row['CO_MUNICIPIO_ESC'][0] == '2':
        return 'NORDESTE'
    elif row['CO_MUNICIPIO_ESC'][0] == '3':
        return 'SUDESTE'
    elif row['CO_MUNICIPIO_ESC'][0] == '4':
        return 'SUL'
    elif row['CO_MUNICIPIO_ESC'][0] == '5':
        return 'CENTRO-OESTE'
    else:
        return 'OUTRA'
```

```python
df_brasil_clean['MEDIA_OBJETIVAS'] = df_brasil_clean.apply(media, axis=1)
df_brasil_clean['REGIAO'] = df_brasil_clean.apply(regiao, axis=1)
```

```python
def intervConfianca(x):
    print("Intervalo de Confiança ao nível de 95% para média quando a variância é desconhecida")
    inf,sup = norm.interval(0.95, st.mean(x), st.stdev(x)/np.sqrt(len(x)))
    print ('Limite inferior 95%:', round(inf,3))
    print ('Limite superior 95%:', round(sup,3))
```

## Teste ANOVA

Hipóteses

H0: As médias são iguais entre os diversos níveis de escolarida da mãe.

H1: As médias são diferentes entre os diversos níveis de escolarida da mãe.

Obs. Assumiu-se distribuição normal e variâncias semelhantes.

```python
lm_escolaridade = ols('NU_NOTA_REDACAO ~ (Q002)', df_manaus).fit()
sm.stats.anova_lm(lm_escolaridade,typ=2)
```

P-value é 3.262544e-226, que é menor que alpha=0.05. Então rejeita-se H0, e aceita H1, as médias são estatisticamente diferentes.
