# Importação de bibliotecas

```
import pandas as pd
from google.colab import files
from restcountries import RestCountryApiV2 as rapi
```

# Dataset0: https://restcountries.eu/

```
all_countries = rapi.get_all()
```

```
dataset0_restcountries = pd.DataFrame([country.__dict__ for country in all_countries])
```

```
dataset0_restcountries
```

# Dataset1: flags.csv

Primeiramente, faça *upload* do arquivo ***flags.csv*** para a pasta ***sample_data*** do colab.

```
dataset1_flags = pd.read_csv('sample_data/flags.csv',encoding= 'latin_1',sep=',')
```

```
dataset1_flags['name'] = dataset1_flags['name'].str.replace('-', ' ')
```

```
dataset1_flags
```

# Dataset2: countries.csv

Primeiramente, faça *upload* do arquivo ***countries.csv*** para a pasta ***sample_data*** do colab.

```
dataset2_countries = pd.read_csv('sample_data/countries.csv',encoding= 'latin_1',sep=';',na_values='null',keep_default_na=False)
```

```
dataset2_countries
```

# Integração dos Datasets

```
paises_somente_df_countries = dataset2_countries[dataset2_countries.alpha_2.isin(dataset0_restcountries.alpha2_code)==False]
```

```
paises_somente_df_countries
```

OBS: **Os dois países a mais no dataset 'countries.csv' em relaçao ao dataset 'restcountries' já não existem mais.**

```
paises_somente_df_restcoun = dataset0_restcountries[dataset0_restcountries.alpha2_code.isin(dataset2_countries.alpha_2)==False]
```

```
paises_somente_df_restcoun
```

```
dataset0_e_dataset2 = pd.merge(dataset0_restcountries,dataset2_countries, how="left",left_on="alpha2_code",right_on="alpha_2")
```

```
dataset0_e_dataset2
```

```
all_datasets = pd.merge(dataset0_e_dataset2,dataset1_flags,how="left",left_on="name_x",right_on="name")
```

```
all_datasets
```

# Download para CSV do novo dataset

```
all_datasets.to_csv("new_dataset.csv",index=False)
files.download("new_dataset.csv")
```
